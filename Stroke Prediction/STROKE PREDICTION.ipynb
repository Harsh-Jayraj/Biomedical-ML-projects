```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
stroke = pd.read_csv("healthcare-dataset-stroke-data.csv")
stroke.head(20)
```

```python
stroke.info()
```

```python
stroke.isna().any()
```

```python
stroke.fillna(np.mean(stroke['bmi']),inplace=True)
stroke.head(10)
```

```python
stroke.isna().any()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in stroke.columns:
    if stroke[i].dtype=='object':
        stroke[i]=le.fit_transform(stroke[i])
```

```python
stroke.head(20)
```

```python
stroke=stroke.drop(['id'],axis=1)
stroke=stroke.values
X,y=stroke[:,0:10],stroke[:,10]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler(feature_range=(0,1))
reX_train = scalar.fit_transform(X_train)
reX_test=scalar.transform(X_test)
```

```python
from sklearn import svm
support = svm.LinearSVC(random_state=20)
```

```python
support.fit(reX_train,y_train)
y_pred=support.predict(reX_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
score=accuracy_score(y_test,y_pred)
print("Your Model Accuracy is", score)
print(confusion_matrix(y_test,y_pred))
```
